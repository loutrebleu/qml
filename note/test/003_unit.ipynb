```python
from qiskit import QuantumCircuit
```

```python
from qml.model.unit import UnitManager
```

```python
num_qubits = nq = 3
num_gates = ng = 3
```

```python
uman = UnitManager(nq, ng)
```

```python
unit = uman.generate_random_unit()
```

```python
qc = QuantumCircuit(nq, 1)
qc = unit.apply_to_qc(qc)
qc.draw("mpl")
```

```python
ustr = unit.to_string()
print(ustr)
```

```python
runit = uman.new_from_string(ustr)
qc = QuantumCircuit(nq, 1)
qc = runit.apply_to_qc(qc)
qc.draw("mpl")
```
