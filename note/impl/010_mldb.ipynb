```python
from qml.db import target as xtarget
from qml.tools.dataset import Dataset
from qml.tools.random import XRandomGenerator
```

```python
class MLData:

    def __init__(self):
        pass
```

```python
class MLDatasetGenerator:

    def __init__(self, function_generator: xtarget.TargetFunctionGenerator, seed: int = None):
        self.rng = XRandomGenerator(seed)
        self.fgen = function_generator
    
    def generate(self, size: int):
        func = self.fgen.generate()
        xs = self.rng.uniform(-1, 1, size)
        ys = func(xs)
        return Dataset(xs, ys)
```

```python
dgen = MLDatasetGenerator(xtarget.PolynominalTargetFunctionGenerator(5))
dataset = dgen.generate(30)

plt.scatter(dataset.xs, dataset.ys)
```
